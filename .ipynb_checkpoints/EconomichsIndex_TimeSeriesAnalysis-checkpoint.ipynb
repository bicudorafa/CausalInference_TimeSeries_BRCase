# Google's Causal Impact Package - BR use case

Making causal inferences is much of the time a complex exercice taking in consideration the difficulties in isolating causal effects or correlations among treateds, non-treateds and covariates. Aiming to execute this kind of analysis, there are huge numbers of frameworks with different approachs to achieve this goal. Here, I'm going to explore the one made to deal with studies in which there isn't available control subjects to make inferences: synthetic conter-factual analysis.

The [Causal Impact](https://storage.googleapis.com/pub-tools-public-publication-data/pdf/41854.pdf) is a framework developed by Google with the following procedure: given a response time series (e.g., clicks) and a set of control time series (e.g., clicks in non-affected markets or clicks on other sites), the package constructs a Bayesian structural time-series model. This model is then used to try and predict the counterfactual, i.e., how the response metric would have evolved after the intervention if the intervention had never occurred.

[Causalimpact](https://github.com/dafiti/causalimpact) is a Python implementation of Google's [R Package](https://google.github.io/CausalImpact/CausalImpact.html) with all functionalities fully ported and tested.

### Assumptions

The main assumption is that there is a set control time series that were themselves not affected by the intervention. If they were, we might falsely under- or overestimate the true effect. Or we might falsely conclude that there was an effect even though in reality there wasnâ€™t. The model also assumes that the relationship between covariates and treated time series, as established during the pre-period, remains stable throughout the post-period.

# Pratical Example - BR Stock Index

Aiming to create a practical example of the package usage, I'll develop a study based on the BR main stock index: Ibovespa (IBOV), during an event in which we have huge media coverage and high drop in the index price: JBS's CEO released an audio involving the actual president, Michel Temer, in corruption. The objective will be to infer if this event really had an causal impact on the stock price and, if it had, how strong it was

As covariates, I selected other financial time series based on these assumptions: 
- US industrial index (DJI): overall global markets trend 
- Shanghai Composite Index (SSE): emerging markets trend
- Commodity Research Bureau Index (CRB): Brazil high sensibility to commodities prices

# Importing Dependencies and Functions

```python
# data analysis
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline 
plt.rcParams['figure.figsize'] = (15, 6)

# python utils
import gc
gc.enable()
import warnings
warnings.filterwarnings('ignore')

# Statistical packages - to download CausalImpact: !pip install pycausalimpact
from causalimpact import CausalImpact
from sklearn import preprocessing
```

```python
def df_scaler(df,scaler):
    """Function to quickly scale a df and return it into the same pd df shape
    scaler must be a sklearn preprocessing object, initiated as parameter"""
    init_scaler=scaler
    scaled_matrix=init_scaler.fit_transform(df)
    df_scaled=pd.DataFrame(scaled_matrix, columns = df.columns, index=df.index)
    return(df_scaled)
```

# Raw Data Ingestion

```python
ibov = pd.read_csv('^BVSP.csv')
print(ibov.dtypes)
ibov.head()
```

```python
dji = pd.read_csv('^DJI.csv')
print(dji.dtypes)
dji.head()
```

```python
sse = pd.read_csv('SSE.csv')
print(sse.dtypes)
sse.head()
```

```python
crb = pd.read_csv('CRBQX.csv')
print(crb.dtypes)
crb.head()
```

### Data Cleaning

To prepare the data for the analysis, I proceded with the creation of a dataset merging all information, series_df, and applying a log transformation aiming to reduce the strong volatility in daily stock prices

```python
# loop to coerce all Date columns to datetime format and set it to index
series=[ibov,dji,sse,crb] 
for serie in series:
    serie['Date'] = pd.to_datetime(serie['Date'])
    serie.set_index('Date', inplace=True)
```

```python
# pipe to join all the data, rename it, and let only full observations
series_df = (
    ibov
    .loc[:,['Close']]
    .join(dji.loc[:,['Close']], rsuffix='DJI') # necessary fisrt rename to not crash pipe
    .join(sse.loc[:,['Close']], rsuffix='SSE')
    .join(crb.loc[:,['Price']], rsuffix='CRB')
    .rename(columns={'Close':'IBOV','CloseDJI':'DJI','CloseSSE':'SSE','Price':'CRB'}) # name coercer to better identify
    .dropna()
    .apply(lambda x: np.log(x))
)
# sanity check
series_df.head()
```

# Data Analysis

Below, it's possible to see how or initial assumption has been confirmed: the series are correlated. More than this, it's possible to see that the series owns a seasonal effect in a weekly basis, and it still has a instable behavior even with the log transformation

```python
# min max scaler to look at the series trend at the same level
series_df.pipe(df_scaler,scaler=preprocessing.MinMaxScaler()).plot();
# person corr to check or assumption of resonable correlation
series_df.corr()
```

```python
seasons=(
    series_df
    .reset_index()
    .assign(
        #dayweek=lambda x:x.Date.apply(lambda x:x.weekday())
        yearweek=lambda x:x.Date.apply(lambda x:x.week)
    )
)
sns.boxplot(x="yearweek", y="IBOV", data=seasons)
plt.title('IBOV distribution through days of the week');
```

# Causal Analysis

Finally, I'll proceed to the causal analysis. To better understand the difference between using the series by itself or along with its covariates, both will be executed More than this, worth mentioning the following:
- Financial series own a lot of volatility, then I'll use just 1 year and a half to train the model
- Due to the above as well, I followed the recommendation in Google's Oficial documentation about parameter "prior_level_sd" recommended to be set to "1" (Check the documentation for a better explanation)
- When using this specific parameter configuration, this may sometimes give rise to unrealistically wide prediction intervals, thus imposing making 1 month forecasting

The final result is summarized at the end by using the specific configuration "output='report'"

```python
joesleyAudios ='20170518'
pre_period = ['20160104', '20170517']
post_period = [joesleyAudios, '20170619']
```

```python
# analysis setting
ci = CausalImpact(
    series_df['IBOV'], pre_period, post_period
    ,prior_level_sd=0.1,niter=1000,nseasons=[{'period': 53}]
)
# plot and report
print(ci.summary())
ci.plot()
```

```python
# analysis setting
ci = CausalImpact(
    series_df, pre_period, post_period
    ,prior_level_sd=0.1,niter=1000,nseasons=[{'period': 53}]
)
# plot and report
print(ci.summary())
ci.plot()
```

```python
print(ci.summary(output='report'))
```

# Conclusions and Thoughts

This analysis was a good exercice on applying a synthetic counter-factual methodology to infer causal effect, however it has it own issues that worth mentioning here for future implementations:
- The original methodology was built for R usage, then this might lead to some different results when using it on both languages, and this should lead to extra attention at the results
- This methodology has its own methodologies when applied to series with high variation. May be interesting to search for a more specific framework for this situations
